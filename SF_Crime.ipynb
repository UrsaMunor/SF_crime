#### Q3 question (OLAP)
Count the number of crimes each "Sunday" at "SF downtown".   
SF downtown is defiend  via the range of spatial location. I used a rectangle to define the SF downtown
reference: https://changhsinlee.com/pyspark-udf/
rectangle: a < x < b  and c < y < d. thus, San Francisco Latitude and longitude coordinates are: 37.773972, -122.431297. X and Y represents each. So we assume SF downtown spacial range: X (-122.4213,-122.4313), Y(37.7540,37.7740).

```
#Spark SQL based
crimeSunday = spark.sql("""WITH crimeSunday AS(
                           SELECT Date FROM sf_crime
                           WHERE DayOfWeek = 'Sunday' 
                           AND X < -122.4213 
                           AND X > -122.4313 
                           AND Y > 37.7540 
                           AND Y < 37.7740)
                           
                           SELECT Date, COUNT(*) as Count
                           from crimeSunday
                           group by Date
                           order by Date""")
#crimeSunday = spark.sql("SELECT  DayOfWeek, COUNT(*) AS Count FROM sf_crime WHERE DayOfWeek = 'Sunday' AND X < -122.4213 AND X > -122.4313 AND Y > 37.7540 AND Y < 37.7740 GROUP BY DayOfWeek")

display(crimeSunday)
```

```
crimeDistrict = spark.sql("""SELECT  Category, HOUR(Time), COUNT(*) AS Count FROM sf_crime 
                             WHERE PdDistrict = "SOUTHERN" OR PdDistrict = "MISSION" OR PdDistrict = "CENTRAL"
                             GROUP BY Category, Hour(Time)
                             ORDER BY Category, Hour(Time)""")
display(crimeDistrict)
```

#### Step 6 (OLAP)
For different category of crime, find the percentage of resolution. Based on the output, give my hints to adjust the policy.

```
crime7 = spark.sql("""SELECT Category, COUNT(
                      CASE WHEN resolution <> 'NONE' THEN 1 ELSE NULL END)/COUNT(*) 
                      AS Percentage FROM sf_crime 
                      GROUP BY Category 
                      ORDER BY Percentage DESC""")
display(crime7)
```
